```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
abalone = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data')
```

```python
abalone.columns = ['Sex',
                   'Length',
                   'Diameter',
                   'Height',
                   'Whole_weight',
                   'Shucked_weight',
                   'Viscera_weight',
                   'Shell_weight',
                   'Rings']
```

```python
abalone.head(10)
```

```python
sns.swarmplot(x='Sex', y='Length', data=abalone)
plt.xlabel('Sex')
plt.ylabel('Length')
plt.show() # 1 continuous variable and 1 categorical variable plotted 1 of 6 ways
```

```python
sns.factorplot(x='Sex', y='Length', data=abalone, kind='bar')
plt.show() # 2 of 6 
```

```python
sns.violinplot(x='Sex', y='Length', data=abalone)
plt.show() # 3 of 6
```

```python
sns.boxplot(x='Sex', y='Length', data=abalone)
plt.show() # 4 of 6
```

```python
sns.lvplot(x='Sex', y='Length', data=abalone)
sns.stripplot(x='Sex', y='Length', data=abalone, size=1, jitter=True, color='black')
plt.show() # 5 of 6
```

```python
sns.pointplot(x='Sex', y='Length', data=abalone)
plt.show() # 6 of 6 ways
```

```python
sns.barplot(x='Sex', y='Length', data=abalone)
plt.show() # 7 of 6 ways
```

```python
plt.hist(abalone['Whole_weight'])
plt.show() # plot 1 variable 4 ways
```

```python
sns.kdeplot(abalone['Whole_weight'])
plt.show() # 2 of 4
```

```python
sns.boxplot(abalone['Whole_weight'])
plt.show() # 3 of 4
```

```python
sns.kdeplot(abalone['Whole_weight'], shade=True)
sns.rugplot(abalone['Whole_weight'])
plt.show() # 4 of 4
```

```python
sns.lmplot(x='Length', y='Shucked_weight', data=abalone, fit_reg=False)
plt.show()

sns.lmplot(x='Length', y='Shucked_weight', data=abalone, fit_reg=True)
plt.show() # plot 2 continuous variables 1 of 3 ways
```

```python
sns.kdeplot(abalone['Length'], abalone['Shucked_weight'], shaded=True)
plt.show() # 2 of 3 ways
```
